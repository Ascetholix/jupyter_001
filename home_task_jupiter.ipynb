```python
from sympy import *
from sympy.plotting import plot 
x = Symbol('x')
f = 0.6*x**3 + 5.5*x**2 +10*x -5
a = 0.6
b = 5.5
c = 10
d =-5
f
```

```python
x1=solve(f)[0]
x2=solve(f)[1]
x3=solve(f)[2]
print(f"Корень x1 = {round(x1,3)}")
print(f"Корень x2 = {round(x2,3)}")
print(f"Корень x3 = {round(x3,3)}")
plot(f)
xmax = -5
ymax = 0.6*(xmax**3) + 5.5*(xmax**2) +10*xmax -5
xmin = -b/3*a
ymin = 0.6*(xmin**3) + 5.5*(xmin**2) +10*xmin -5
print(f'Точка max = {xmax,ymax}')
print(f'Точка min = {round(xmin,3),round(ymin,3)}')
```
