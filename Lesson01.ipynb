```python
from sympy import *
from sympy.plotting import plot 
x = Symbol('x')
f = 5*x**2 + 10*x -30
print(solve(f))
x1 =solve(f)[0]
x2 =solve(f)[1]
print(float(x1))
print(float(x2))
plot(f)
```

```python
a = 5
b = 10
c = -30
xmin = -(b)/(2*a)
print(f"Функция возрастает на промежутке от {xmin} до бесеонечности")
print(f"Функция убывает на промежутке от  -бесеонечности до {xmin}")
```

```python
y = 5*(xmin*xmin)+(b*xmin)+(c)
print(xmin,y)
print(f"Функция положительна на промежутке от - бесконечности до ({x1}) и от ({x2}) до + бесконечности")
print(f"Функция отрицательна от ({x1}) до ({x2}) ")
```
